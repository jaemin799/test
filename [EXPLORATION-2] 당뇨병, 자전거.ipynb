```python
# 가중치(W)와 편향(b) 구하기
# 참고 EXPLORATION[2- 8]

# 가중치는 데이터 속성 갯수만큼 구해야 합니다.
W = np.random.rand(10)

# 편향은 하나만 써야 하기 때문에 하나만 구합니다.
b = np.random.rand()

print("-----------------가중치-------------")
print(W)
print("-----------------편향----------------")
print(b)
```

```python
#모델 구하기
#참고 EXPLORATION[2-8]
#회고 - 입력값 * 가중치 + 편향을 원리로 모델 함수를 만드는 거 같음
#회고 - predictions은 x * w + x * w + b를 성립하기 위한 변수
#회고 x[:,i] 왜 이렇게 되는가 한참 생각하다가 x_train.shape를 보고 짐작이 갔음
def model(X, W, b):
    predictions = 0
    for i in range(10):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions
```

```python
# MSE  정의하기
#참고 EXPLORATION[2-8]
#회고 - MSE의 계산식을 까먹음. 다시 복습해야 되겠음


def MSE(a, b):
    mse = ((a - b) ** 2).mean() 
    return mse
```

```python
# 손실함수 만들기
# 참고 EXPLORATION[2-8]

def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
```

```python
# 기울기 구하기
# 참고 EXPLORATION[2-9]

def gradient(X, W, b, y):
    N = len(W)
    
    y_pred = model(X, W, b)
    
    dW = 1/N * 2 * X.T.dot(y_pred - y)
        
    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
# 하이퍼 파라미터 학습률 설정하기
# 자동화 하기 위해서 while문에 for문 선언하기 검색해보자
#이중 for문으로 해보려고 했으나 실패

#learning_rate = np.arange(0.0001, 1, 0.001)

#for i in learning_rate:
#    for j in range(1, 1001):
#        dw, db = gradient(x_train, w, b, y_train)
#        w -= i * dw
#        b -= i * db
#        L = loss(x_train, w, b, y_train)
#        losses.append(L)
#        if i % 10 == 0:
#            print('Iteration %d : Loss %0.4f' %(j, L))
#        if L < 3000:
#            break
```

```python
# 자동화 하이퍼 파라미터 코딩 실패해서 일일이 쳐서 확인했어요.
# 지금은 실패했지만 나중에 다시 도전했을 때 자동화 하이퍼 파라미터 코딩 성공
# 하겠습니다.

LEARNING_RATE = 0.1

losses = []

for i in range(1, 1001):
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 10 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
# 참고 EXPLORATION[2-10]
# 간단한 코딩인데 이걸 참고해서 수행을 한다는 게 나의 부족함을 알게 되었음

plt.plot(losses)
plt.show()
```
